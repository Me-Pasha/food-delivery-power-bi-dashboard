**time_orderd_picked**

```python
df["Time_Order_picked"].value_counts() #doesn't seem any null values
```

```python
df["Time_Order_picked"].unique()
```

```python
df['Time_Order_picked'] = pd.to_datetime(
    df['Time_Order_picked'],
    format='%H:%M:%S',
    errors='coerce'
)
```

```python
order_mins = (
    df["Order_Date_Time"].dt.hour * 60 +
    df["Order_Date_Time"].dt.minute
)

pickup_mins = (
    df["Time_Order_picked"].dt.hour * 60 +
    df["Time_Order_picked"].dt.minute
)

df["pickup_delay_mins"] = pickup_mins - order_mins
```

```python
df.drop(df[df["pickup_delay_mins"] < 0].index, inplace=True) 
```

```python
df["pickup_delay_mins"].unique()
```

```python
df.drop("Time_Order_picked", axis=1, inplace=True)
```
